### Note: Caution using importance scores in XGBoost

Why do we need the less important features, isn't the idea that less is more in machine learning?

To answer this lets use the top 3 to see what happens


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
```

```python
data = pd.read_csv("data/titanic.csv")
```

```python
# Select our features
data = data[['Sex', 'Survived', 'Parch', 'SibSp']]
```

```python
data.head(20)
```

```python
data['Sex'] = data['Sex'].map({'male': 0, 'female': 1})
```

```python
# looking for NaN
data.head(25)
```

```python
# drop all missinng values in all columns
data = data.dropna()
```

```python
X = data.drop('Survived', axis=1)
y = data['Survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
model = XGBClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100))
```
